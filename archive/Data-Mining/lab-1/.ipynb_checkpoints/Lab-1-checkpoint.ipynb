```python
import pandas as pd
import numpy as np

import glob, os
```

```python
path = os.path.dirname(os.path.realpath( './laboratory-work-1' ))
files = glob.glob(path + "/*.txt")
frame = pd.DataFrame()

data = dict({})
sample = []

for file in files:
    parts = file.split("_")
    values = data.get(parts[1], [])
    values.append(pd.read_csv(file,index_col=None, header=0, names=['Date', parts[1]] ))
    data.update({ parts[1] : values })

usd = pd.concat(data.get("au"), sort=True)
eur = pd.concat(data.get("euro"), sort=True)
```

```python
eur.merge(usd, how='left', on="Date")
```
